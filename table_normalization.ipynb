```python
import pandas as pd
import numpy as np
```

```python
# Leemos en datasets beta
bitcoin_price = pd.read_csv("data/beta/bitcoin_price.csv")
ethereum_price = pd.read_csv("data/beta/ethereum_price.csv")
ripple_price = pd.read_csv("data/beta/ripple_price.csv")
solana_price = pd.read_csv("data/beta/solana_price.csv")
okb_price = pd.read_csv("data/beta/okb_price.csv")
arbitrum_price = pd.read_csv("data/beta/arbitrum_price.csv")
optimism_price = pd.read_csv("data/beta/optimism_price.csv")
elrond_price = pd.read_csv("data/beta/elrond_price.csv")
boba_price = pd.read_csv("data/beta/boba_price.csv")
gmd_price = pd.read_csv("data/beta/gmd_price.csv")
```

```python
bitcoin_halvings = pd.read_csv("data/beta/bitcoin_halvings.csv")
```

```python
bitcoin_price['timestamp'] = pd.to_datetime(bitcoin_price['timestamp'])
ethereum_price['timestamp'] = pd.to_datetime(ethereum_price['timestamp'])
ripple_price['timestamp'] = pd.to_datetime(ripple_price['timestamp'])
solana_price['timestamp'] = pd.to_datetime(solana_price['timestamp'])
okb_price['timestamp'] = pd.to_datetime(okb_price['timestamp'])
arbitrum_price['timestamp'] = pd.to_datetime(arbitrum_price['timestamp'])
optimism_price['timestamp'] = pd.to_datetime(optimism_price['timestamp'])
elrond_price['timestamp'] = pd.to_datetime(elrond_price['timestamp'])
boba_price['timestamp'] = pd.to_datetime(boba_price['timestamp'])
gmd_price['timestamp'] = pd.to_datetime(gmd_price['timestamp'])
bitcoin_halvings['timestamp'] = pd.to_datetime(bitcoin_halvings['timestamp'])
```

```python
ethereum_price.head()
```

```python
# Crear una tabla de fechas Ãºnica
unique_dates = pd.concat([bitcoin_halvings['timestamp'],
                          arbitrum_price['timestamp'],
                          bitcoin_price['timestamp'],
                          boba_price['timestamp'],
                          elrond_price['timestamp'],
                          ethereum_price['timestamp'],
                          gmd_price['timestamp'],
                          okb_price['timestamp'],
                          optimism_price['timestamp'],
                          ripple_price['timestamp'],
                          solana_price['timestamp']]).unique()

df_date = pd.DataFrame({'date': unique_dates})
df_date = df_date.sort_values(['date'], ignore_index=True)
df_date['date_id'] = range(1, len(df_date) + 1)
df_date['year'] = df_date['date'].dt.year
df_date['month'] = df_date['date'].dt.month
df_date['day'] = df_date['date'].dt.day
```

```python
df_date.head()
```

```python
df_date.to_csv("data/launch/crypto_date.csv", index=False)
```

```python
# Unificar todas las tablas de precios
all_prices = pd.concat([arbitrum_price, bitcoin_price, boba_price, elrond_price, ethereum_price, gmd_price, okb_price, optimism_price, ripple_price, solana_price])
merged_prices = all_prices.merge(df_date[["date", "date_id"]], left_on='timestamp', right_on='date', how='inner')
merged_prices = merged_prices.drop(columns=['date'])
merged_prices = merged_prices.sort_values(['timestamp'], ignore_index=True)
merged_prices = merged_prices[['cryptocurrency_id', 'date_id', 'price_btc', 'price_eth', 'price_eur', 'price_usd', 'price_satoshis', 'market_cap_btc', 'market_cap_eth', 'market_cap_eur', 'market_cap_usd', 'market_cap_satoshis', 'total_volume_btc', 'total_volume_eth', 'total_volume_eur', 'total_volume_usd', 'total_volume_satoshis']]
```

```python
merged_prices.head()
```

```python
merged_prices.to_csv("data/launch/crypto_price.csv", index=False)
```
